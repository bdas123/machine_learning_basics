```python
# Import modules
import pandas as pd
import numpy as np
import sklearn
import mglearn
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
#Import scikit-learn
from sklearn.model_selection import train_test_split
```

```python
# Read CSV file
df = pd.read_csv("Mall Customers.csv")
df.head()
```

```python
# Separate the columns
gender = df.iloc[:,1]
age = df.iloc[:,2].values.reshape(-1,1)
annual_income = df.iloc[:,3].values.reshape(-1,1)
spending_score = df.iloc[:,4].values.reshape(-1,1)
```

```python
# K-Neighbors Classifier for Age and Spending Score
age_train, age_test, spending_score_train, spending_score_test = train_test_split(age, spending_score, random_state=0)
```

```python
# Create Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_accuracy_age = []
test_accuracy_age = []

#try n-neighbors from 1 to age_train.size

neighbors_settings_age = range(1,age_train.size)
 
for n_neighbors in neighbors_settings_age:
    #build the model
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(age_train, spending_score_train)
    
    #record training set accuracy
    training_accuracy_age.append(clf.score(age_train, spending_score_train))
    
    #record generalization accuracy
    test_accuracy_age.append(clf.score(age_test, spending_score_test))
    
```

```python
plt.plot(neighbors_settings_age, training_accuracy_age, label = "Training Accuracy", c = 'b', linewidth = 6.0)
plt.plot(neighbors_settings_age, test_accuracy_age, label = "Test Accuracy", c = 'r', linewidth = 6.0)
plt.xlabel("Number of Neighbors", fontsize = 14)
plt.ylabel("Accuracy", fontsize = 14)
plt.title("Accuracy vs Number of Neighbors (Age)", fontsize = 16)
plt.legend(fontsize = 13)
```

```python
annual_income_train, annual_income_test, spending_score_train, spending_score_test = train_test_split(annual_income, spending_score, random_state=0)

```

```python
training_accuracy_annual_income = []
test_accuracy_annual_income = []

#try n-neighbors from 1 to age_train.size

neighbors_settings_annual_income = range(1,annual_income_train.size)
 
for n_neighbors in neighbors_settings_annual_income:
    #build the model
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(annual_income_train, spending_score_train)
    
    #record training set accuracy
    training_accuracy_annual_income.append(clf.score(annual_income_train, spending_score_train))
    
    #record generalization accuracy
    test_accuracy_annual_income.append(clf.score(annual_income_test, spending_score_test))
```

```python
plt.plot(neighbors_settings_annual_income, training_accuracy_annual_income, label = "Training Accuracy", c = 'b', linewidth = 6.0)
plt.plot(neighbors_settings_annual_income, test_accuracy_annual_income, label = "Test Accuracy", c = 'r', linewidth = 6.0)
plt.xlabel("Number of Neighbors", fontsize = 14)
plt.ylabel("Accuracy", fontsize = 14)
plt.title("Accuracy vs Number of Neighbors (Annual Income)", fontsize = 16)
plt.legend(fontsize = 13)
```

```python
plt.figure(figsize = (8,5))
plt.plot(neighbors_settings_age, test_accuracy_age, label = "Test Accuracy (Age)", c = 'r', linewidth = 6.0)
plt.plot(neighbors_settings_annual_income, test_accuracy_annual_income, label = "Test Accuracy (Annual Income)", c = 'y', linewidth = 6.0)
plt.xlabel("Number of Neighbors", fontsize = 14)
plt.ylabel("Accuracy", fontsize = 14)
plt.title("Test Accuracy vs Number of Neighbors", fontsize = 16)
plt.legend(fontsize = 13)
```

```python
plt.figure(figsize = (8,5))
plt.plot(neighbors_settings_age, training_accuracy_age, label = "Training Accuracy (Age)", c = 'r', linewidth = 6.0)
plt.plot(neighbors_settings_annual_income, training_accuracy_annual_income, label = "Training Accuracy (Annual Income)", c = 'y', linewidth = 6.0)
plt.plot(neighbors_settings_age, test_accuracy_age, label = "Test Accuracy (Age)", c = 'b', linewidth = 6.0)
plt.plot(neighbors_settings_annual_income, test_accuracy_annual_income, label = "Test Accuracy (Annual Income)", c = 'g', linewidth = 6.0)
plt.xlabel("Number of Neighbors", fontsize = 14)
plt.ylabel("Accuracy", fontsize = 14)
plt.title("Training Accuracy vs Number of Neighbors", fontsize = 16)
plt.legend(fontsize = 13)
```
